```python
def cls_pooling(model_output):
    return model_output.last_hidden_state[:, 0]

def get_embeddings(text_list, imp_model, imp_tokenizer):
    # encoded_input = imp_tokenizer(
    #     text_list, padding=True, truncation=True, return_tensors="pt"
    # )
    encoded_input = imp_tokenizer(
        text_list, return_tensors="pt"
    )
    encoded_input = {k: v.to(device) for k, v in encoded_input.items()}
    model_output = imp_model(**encoded_input)
    return cls_pooling(model_output)
```

## Funcs

### Embed table

```python
def create_embed_table(conn: Connection, table_name: str, embedding_length: str):

    try:
        # load the extention
        conn.enable_load_extension(True) # start loading extensions
        sqlite_vec.load(conn)
        conn.enable_load_extension(True) # end loading extensions

        sql_statement = f"""
        CREATE VIRTUAL TABLE {table_name} USING vec0(
            embedding float[{embedding_length}]);
        """
        cursor = conn.cursor()
        cursor.execute(sql_statement)
        conn.commit()
    except:
        print("Failed to create embedding tables")
```

```python
def add_embeddings(conn: Connection, table_data, table_name: str):
    # insert table statement
    sql = f"INSERT INTO {table_name}(rowid, embedding) VALUES (?, ?)"
    try:
        conn.enable_load_extension(True) # start loading extensions
        sqlite_vec.load(conn)
        conn.enable_load_extension(True) # end loading extensions
        # Create  a cursor
        cur = conn.cursor()

        # execute the INSERT statement
        cur.execute(sql, table_data)

        # commit the changes
        conn.commit()
    except sqlite3.OperationalError as e:
        print("Failed to add datat to embedding table.", e)

    # get the id of the last inserted row
    return cur.lastrowid
```

```python
def delete_embed_table(conn: Connection, table_name: str):
    try:
            conn.enable_load_extension(True) # start loading extensions
            sqlite_vec.load(conn)
            conn.enable_load_extension(True) # end loading extensions

            sql = f"DROP TABLE IF EXISTS {table_name}"
            cursor = conn.cursor()
            cursor.execute(sql)
            conn.commit()
    except sqlite3.OperationalError as e:
        print("Failed to delete to embedding table.", e)
```

```python
def create_table(conn: Connection, meta_table_name: str, col_names: list, col_conditions: list):

    column_info = ""
    total_cols = len(col_names)
    for idx, col in enumerate(col_names):
        if idx < (total_cols -1):
            column_info = column_info + col + " " + col_conditions[idx] + ",\n"
        else:
            column_info = column_info + col + " " + col_conditions[idx]
            
    sql_statement = f"""
    CREATE TABLE IF NOT EXISTS {meta_table_name} (
            {column_info}
        );
    """
    try:
        cursor = conn.cursor()
        cursor.execute(sql_statement)
        conn.commit()
    except sqlite3.OperationalError as e:
        print("Failed to create meta table", e)
```

```python
def add_to_meta_data(conn: Connection, table_data, meta_table_name, which_cols: list):
    # insert table statement

    col_queries = ""
    len_queries = len(which_cols)
    val_queries = ""
    for idx, each_query in enumerate(which_cols):
        if idx < (len_queries-1):
            col_queries = col_queries + each_query + ","
            val_queries = val_queries+ "?" + ","
        else:
            col_queries = col_queries + each_query
            val_queries = val_queries + "?"

    sql = f""" INSERT INTO {meta_table_name}({col_queries})
              VALUES({val_queries}) 
            """
    try:
        # Create  a cursor
        cur = conn.cursor()

        # execute the INSERT statement
        cur.execute(sql, table_data)

        # commit the changes
        conn.commit()
    except sqlite3.OperationalError as e:
        print("Failed to add datat to table.", e)

    # get the id of the last inserted row
    return cur.lastrowid
```

```python
def delete_meta_table(conn: Connection, meta_table_name: str) :
    try:
            sql = f"DROP TABLE IF EXISTS {meta_table_name};"
            cursor = conn.cursor()
            cursor.execute(sql)
            conn.commit()
    except sqlite3.OperationalError as e:
        print("Failed to delete to meta table.")
```
